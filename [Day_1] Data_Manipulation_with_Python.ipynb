## Missing Value

```python
step1 = pd.DataFrame(data.dtypes).rename(columns = {0: 'Columns Type'}) # Membaca tipe data setiap kolom
step2 = pd.DataFrame(data.isnull().sum()).rename(columns = {0: 'Amount of Null Values'}) # Menghitung jumlah data yang kosong
step3 = pd.DataFrame(round(data.isnull().sum()/data.shape[0]*100,2)).rename(columns = {0: 'Percentage of Null Values'}) # Menghitung persentase data yang kosong

cek_data = pd.concat([step1, step2, step3], axis = 1)
cek_data
```

```python
# Melihat distribusi numerik data
data.describe()
```

```python
# Melihat paling banyak values pada kategorikal data
categorical_cols = data.select_dtypes(include=['object']).columns # Mengambil list kolom kategorikal

for col in categorical_cols:
    print(f"\n📊 Top values for column: {col}")

    # Buat pivot table (count frekuensi setiap nilai unik)
    pivot_df = data[col].value_counts().reset_index()
    pivot_df.columns = [col, 'Count']

    # Urutkan dari yang terbesar
    pivot_df = pivot_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

    # Tampilkan hasilnya
    display(pivot_df)
```

```python
# Setup imputer
num_imputer = SimpleImputer(strategy='mean') # rata-rata jika numerik
cat_imputer = SimpleImputer(strategy='most_frequent') # yang paling sering muncul

# Input numerik values
data['Imputer_previous_year_rating'] = num_imputer.fit_transform(data[['previous_year_rating']]).ravel()

# Input category values
data['Imputer_education'] = cat_imputer.fit_transform(data[['education']]).ravel()
```

```python
# Cek data original yang null
data[data['education'].isnull()]
```

## Pivot Table

Question:
1. Berapa banyak karyawan pada setiap department
2. Berapa rata-rata usia karyawan yang dihire berdasarkan recruitment_channel
3. Berapa rata-rata length_of_service karyawan dilihat berdasarkan deparment (baris) dan recruitment_channel (kolom)

```python
# prompt: Berapa banyak karyawan pada setiap department
question1 = data.pivot_table(index='department', values='employee_id', aggfunc='count').reset_index()
question1.rename(columns={'employee_id': 'Total Karyawan'})
```

```python
# prompt: Berapa rata-rata usia karyawan yang dihire berdasarkan recruitment_channel
question2 = data.pivot_table(index='recruitment_channel', values='age', aggfunc='mean').reset_index()
question2.rename(columns={'age': 'Rata-rata Usia'})
```

```python
# prompt: Berapa rata-rata length_of_service karyawan dilihat berdasarkan deparment (baris) dan recruitment_channel (kolom), dan rounding hasil rata-rata
question3 = data.pivot_table(index='department', columns='recruitment_channel', values='length_of_service', aggfunc='mean').round(2)

question3.columns.name = None #Menampilkan baris, kolom, dan values dengan format Tabular Dataframe
question3.reset_index()
```
