```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
data_text = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/refs/heads/main/dataset/News_Title_NLP.csv")
data_text = data_text.iloc[:5,:]

data_text.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#generate the word cloud
combine_word = "".join([char for char in data_text['News Title']])

print("Hasil Combine Word")
display(combine_word)

print("\nHasil Work Token")
word_tokens = word_tokenize(combine_word)
display(word_tokens)
```

```python
#remove useless word in data
stop_words = set(stopwords.words('english'))
display(stop_words)

print("\nHasil Looping")
clean_words = []

for w in word_tokens:
    if w not in stop_words:
        clean_words.append(w)

display(clean_words)
```

```python
clean_words = [w for w in word_tokens if not w in stop_words]
display(clean_words)
```
